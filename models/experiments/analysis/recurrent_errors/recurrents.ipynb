# Analysis of results

```python
%config IPCompleter.greedy=True
import pandas as pd
from subprocess import call
```

```python
misclass1 = pd.read_csv('../../02-resNet50_pretrained_species/misclassified.csv') 
```

```python
misclass2 = pd.read_csv('../../05-resNet50_pretrained_nofc_species/misclassified.csv') 
```

```python
misclass3 = pd.read_csv('../../10-resNet50_pretrained_nofc_species_inputsize_416/misclassified.csv') 
```

```python
call(["mkdir", "recurrent_img"])
```

```python
recurrents = pd.merge(misclass1, misclass2, on=['Filename'], how='inner')
del recurrents["Unnamed: 0_x"]
del recurrents["Unnamed: 0_y"]
recurrents = pd.merge(recurrents, misclass3, on=['Filename'], how='inner')
del recurrents["Unnamed: 0"]
```

```python
recurrents
```

### Copy images of recurruntly misclassified bettles to folder "recurrent_img"

```python
for name in recurrents["Filename"]:
    im_name = name.split("/")[-1]  
    call(["cp", "../../../../images/labeled_val/"+im_name, "./recurrent_img/"])
```
