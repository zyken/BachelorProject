```python
from PIL import Image
import os
import pandas as pd
from imutils import paths
```

```python
#analysis on shuffled, at this is only recent exp with classification report as .csv
image_dir_shuffled = "../../../../images/images_species_shuffled/train/"
image_dir_normal = "../../../../images/images_species/train/"
```

```python
species_dirs = [os.path.join(image_dir_normal, name) for name in os.listdir(image_dir_normal)]
```

```python
species_dirs
```

```python
list(paths.list_images('../../../../images/images_species_shuffled/train/Leptacinus pusillus'))
```

```python
# get avg pixels pr. species
species_pixels = []
for species_dir in species_dirs:
    species = species_dir.split("/")[-1]
    img_paths = list(paths.list_images(species_dir))
    img_pixels = []
    for path in img_paths:
        img = Image.open(path)
        width,height = img.size
        img_pixels.append(width*height)
    avg_pixels = int(sum(img_pixels) / len(img_pixels))
    species_pixels.append([species, avg_pixels])
species_pixels
```

```python
df = pd.DataFrame(species_pixels,columns=['class','avg_pixels'])
df.to_csv("normal_train_pixels_pr_species.csv")
```
