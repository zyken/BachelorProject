```python
def get_bar(name, t, missclassified, path):
    mc_pr_name = misclassified[misclassified["Label"] == name]
    mc_pr_name = mc_pr_name[["Prediction3a","Prediction3a_prob", "Prediction3b", "Prediction3b_prob", "Prediction3c", "Prediction3c_prob"]]
    a = mc_pr_name["Prediction3a"][mc_pr_name["Prediction3a_prob"] > t]
    b = mc_pr_name["Prediction3b"][mc_pr_name["Prediction3b_prob"] > t]
    c = mc_pr_name["Prediction3c"][mc_pr_name["Prediction3c_prob"] > t]
    x = []
    for v in a.values:
        x.append(v)
    for v in b.values:
        x.append(v)
    for v in c.values:
        x.append(v)
    beat, coun = np.unique(x, return_counts=True)
    plt.figure()
    plt.bar(beat, coun)
    plt.title("misspredictions for " + name )
    plt.xticks(beat, rotation='vertical')
    plt.savefig(path)
    return 
    
    
```

```python
call(["mkdir", "charts_miss"])

for ins in out:
    get_bar(ins, 0.2, misclassified, 'charts_miss/' + ins +"_missmatch.png")



```
