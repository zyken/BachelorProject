# Analysis of results

```python
%config IPCompleter.greedy=True
import pandas as pd
from subprocess import call
```

```python
results = pd.read_csv('results.csv') 
```

```python
call(["mkdir", "misclassed_img"])
```

```python
misclassified = results.loc[results["Label"] != results["Prediction1"]]
del misclassified["Unnamed: 0"]
misclassified.to_csv("misclassified.csv")
```

```python
misclassified
```

### Copy images of misclassified bettles to folder "misclassed_img"

```python
for name in misclassified["Filename"]:
    im_name = name.split("/")[-1]  
    call(["cp", "../../images/labeled_val/"+im_name, "./misclassed_img/"])
```
