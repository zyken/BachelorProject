```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from random import randint
import cv2 as cv
```

```python
def random_blur(im):
    k = randint(0,3)
    #kernels = [3, 5, 13]
    if k == 0:
        return im
    print(k)
    return cv.medianBlur(im, 3)#kernels[k-1])
```

```python
data_gen = ImageDataGenerator(
    rescale=1./255.,
    #Flipping
    horizontal_flip=True,
    vertical_flip=True,
    #translation
    width_shift_range=0.1,
    height_shift_range=0.1,
    #Rotation
    rotation_range=90,
    # Scale
    zoom_range=0.1,
    #Blur
    preprocessing_function=random_blur,
    #Intensity
    brightness_range=[0.3, 1.0]
)
    
```

```python
d = 'testimg/'
iterator = data_gen.flow_from_directory(d,target_size=(225, 225),
        batch_size=1)
im = iterator.next()
im = im[0]
plt.imshow(im[0,:])
    
```
