```python
data_gen = ImageDataGenerator(
    rescale=1./255.,
    #Flipping
    horizontal_flip=True,
    vertical_flip=False,
    #translation
    width_shift_range=0.025,
    height_shift_range=0.025,
    #Rotation
    rotation_range=20,
    # Scale
    zoom_range=0.0,
    #Blur
    #preprocessing_function=random_blur,
    #Intensity
    brightness_range=[0.6, 1.]
)
    
```

```python
d = 'testimg/'
iterator = data_gen.flow_from_directory(d,target_size=(225, 225),
        batch_size=1)
im = iterator.next()
im = im[0]
plt.imshow(im[0,:])
plt.show()
```

```python
accs = [0.87855, 0.88712, 0.89214]
plt.figure()
plt.plot([224, 320, 420], accs)
plt.xlabel('input size')
plt.ylabel('validation accuracy')
plt.xticks([224, 320, 420])
plt.yticks(accs)
plt.title('Validation Accuracy vs input size')
plt.savefig('')
plt.show()
```
